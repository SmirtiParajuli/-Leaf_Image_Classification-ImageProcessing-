```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

def load_file():
    root = tk.Tk()
    root.withdraw() # Hide the main window of Tkinter to only show the file dialog
    # Open a file dialog to select a file with specified file types
    file_path = filedialog.askopenfilename(title="Select File", filetypes=[("All files", "*.*"), ("Excel files", "*.xlsx *.xls"), ("CSV files", "*.csv"), ("XML files", "*.xml")])
    if file_path:# Check if a file path was selected
        try:
             # Load the data into a DataFrame based on the file extension
            if file_path.endswith('.xlsx') or file_path.endswith('.xls'):
                 df_original = pd.read_excel(file_path, engine='openpyxl')
            elif file_path.endswith('.csv'):
                df_original = pd.read_csv(file_path)
            elif file_path.endswith('.xml'):
               df_original = pd.read_xml(file_path)
            else:
                print("Unsupported file type.")
                return None

            print("Successfully read the file.\n")
            print("File content:")
            print( df_original.info())# Display information about the DataFrame
            return  df_original
        except Exception as e:
            print(f"Failed to read the file: {e}")# Print the exception if file loading fails
    else:
        print("No file selected.")
    return None

# Load the dataset
df_original = load_file()

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def split_dataset(df, test_size=0.2):
    # Split the dataset into training and testing sets, stratified by the 'Class' column
    train_df, test_df = train_test_split(df, test_size=test_size, stratify=df['Class'], random_state=42)
    
    # Define the filenames for the train and test CSV files
    train_file = 'FeatureTrain.csv'
    test_file = 'FeatureTest.csv'
    
    # Save the train and test sets to CSV files
    train_df.to_csv(train_file, index=False)
    test_df.to_csv(test_file, index=False)
    
    # Print success message
    print(f"Train file '{train_file}' and test file '{test_file}' created successfully.")
    
    # Return the training and testing DataFrames
    return train_df, test_df

def load_data(train_file='FeatureTrain.csv', test_file='FeatureTest.csv'):
    # Load the training and testing data from CSV files
    try:
        feature_train = pd.read_csv(train_file)
        feature_test = pd.read_csv(test_file)
    except FileNotFoundError as e:
        # Print error message if files are not found
        print(f"Error: {e}")
        return None
    
    # Separate features and target variable for training data
    X_train = feature_train.drop(columns='Class')
    y_train = feature_train['Class']
    
    # Separate features and target variable for test data
    X_test = feature_test.drop(columns='Class')
    y_test = feature_test['Class']
    
    # Print the number of samples in the training and testing sets
    print("Number of samples in training set:", feature_train.shape[0])
    print("Number of samples in test set:", feature_test.shape[0])  
    
    # Return the separated features and target variables for training and testing data
    return X_train, y_train, X_test, y_test

def preprocess_data(X_train, X_test):
    # Initialize a StandardScaler
    scaler = StandardScaler()
    
    # Fit the scaler on the training data and transform it
    X_train_scaled = scaler.fit_transform(X_train)
    
    # Transform the test data using the same scaler
    X_test_scaled = scaler.transform(X_test)
    
    # Return the scaled training and testing data
    return X_train_scaled, X_test_scaled

# Call the split_dataset function and get the train and test DataFrames
train_df, test_df = split_dataset(df)

# Call the load_data function to load and separate the data into features and target variables
X_train, y_train, X_test, y_test = load_data()
if X_train is not None and X_test is not None:
    # Call the preprocess_data function to scale the features
    X_train_scaled, X_test_scaled = preprocess_data(X_train, X_test)
    
    # Print success message for preprocessing
    print("\nPreprocessing complete.")
    
    # Optionally, print the shape of the scaled data to verify
    print("Shape of scaled training data:", X_train_scaled.shape)
    print("Shape of scaled test data:", X_test_scaled.shape)
else:
    # Print error message if data loading failed
    print("Data loading failed. Please check the file paths and try again.")
```

```python
from imblearn.over_sampling import SMOTE
# Apply SMOTE to handle class imbalance in the training set
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Preprocess the data
X_train_scaled, X_test_scaled = preprocess_data(X_train_balanced, X_test)
```

```python
import time
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import cross_val_score
from tabulate import tabulate

# Function to calculate specificity
def calculate_specificity(y_true, y_pred):
    # Compute the confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    if cm.shape == (2, 2):  # Binary classification
        # Extract true negatives (tn), false positives (fp), false negatives (fn), and true positives (tp)
        tn, fp, fn, tp = cm.ravel()
        # Calculate specificity for binary classification
        return tn / (tn + fp) if (tn + fp) != 0 else 0.0
    else:  # Multiclass classification
        specificities = []
        for i in range(cm.shape[0]):
            # Calculate true negatives and false positives for each class
            tn = cm.sum() - (cm[i, :].sum() + cm[:, i].sum() - cm[i, i])
            fp = cm[:, i].sum() - cm[i, i]
            # Calculate specificity for each class and append to the list
            specificities.append(tn / (tn + fp) if (tn + fp) != 0 else 0.0)
        # Return the average specificity across all classes
        return sum(specificities) / len(specificities) if specificities else 0.0

# Define models to be evaluated
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),  # Logistic Regression model
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),  # Random Forest model
    "SVM": SVC(kernel='linear', random_state=42),  # Support Vector Machine model with linear kernel
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),  # K-Nearest Neighbors model with 5 neighbors
}
```

```python
# Initialize an empty list to store results
results = []

# Train and evaluate models on the test data, and collect results
for name, model in models.items():
    try:
        start_time = time.time()  # Record the start time

        # Train the model on the scaled training data
        model.fit(X_train_scaled, y_train_balanced)

        # Make predictions on the scaled test data
        test_pred = model.predict(X_test_scaled)

        # Calculate the elapsed time for training and prediction
        elapsed_time = time.time() - start_time

        # Calculate the accuracy on the test data
        test_accuracy = accuracy_score(y_test, test_pred)
        # Calculate the precision on the test data (weighted average)
        test_precision = precision_score(y_test, test_pred, average='weighted')
        # Calculate the recall on the test data (weighted average)
        test_recall = recall_score(y_test, test_pred, average='weighted')
        # Calculate the F1 score on the test data (weighted average)
        test_f1 = f1_score(y_test, test_pred, average='weighted')
        # Calculate the specificity on the test data
        specificity = calculate_specificity(y_test, test_pred)

        # Append the calculated metrics to the results list
        results.append({
            "Model": name,
            "Accuracy": test_accuracy,
            "Precision": test_precision,
            "Recall": test_recall,
            "F1 Score": test_f1,
            "Specificity": specificity,
            "Elapsed Time (s)": elapsed_time
        })

        # Perform cross-validation on the training data and print the mean and standard deviation of accuracy scores
        scores = cross_val_score(model, X_train_scaled, y_train_balanced, cv=5, scoring='accuracy')
        print(f"Cross-validation scores for {name}: {scores.mean()} ± {scores.std()}\n")

    except Exception as e:
        # Print an error message if there's an issue evaluating the model
        print(f"Error evaluating {name}: {e}")

# Convert the results list to a DataFrame
results_df = pd.DataFrame(results)

# Print the results in a tabular format using the tabulate library
print(tabulate(results_df, headers='keys', tablefmt='simple'))

```
